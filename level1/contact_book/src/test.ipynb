```python
from collections import defaultdict
from tabulate import tabulate

class ContactBook():
    def __init__(self):
        self.contact = defaultdict(dict)

    def add_contact(self, name: str, phone: str, email: str = None):
        if name in self.contact:
            print("Contact Already Existed!")
            return
        self.contact[name]["phone"] = phone
        self.contact[name]["email"] = email
        
    def view_contact(self):
        for name, info in self.contact.items():
            print(tabulate([[name, info["phone"], info["email"]]], headers=['Names', 'phones', 'emails']))
            print("-" * 50)
            
    def delete_contact(self, name):
        if name in self.contact:
            del self.contact[name]
            print("Contact deleted successfully!")
        else:
            print("Contact not found!")
    
    def updated_contact(self, name, phone=None, email=None):
        if name in self.contact:
            if phone:
                self.contact[name]['phone'] = phone
            if email:
                self.contact[name]['email'] = email     
            print("Contact updated successfully!")
        else:
            print("Contact not found!")
    
        

            
        

```

```python
book = ContactBook()
```

```python
book.add_contact('ali','99')
```

```python
book.view_contact()
```

```python
book = ContactBook()


while True:
    print("\n\nWelcome to contact book app")
    print("1. Add Contact")
    print("2. Edit Contact")
    print("3. View Contact")
    print("4. Delete contact")
    print("5. Exit")
    
    input_from_user = input("Please choose an option: ")
    if input_from_user == "5":
        break
    
    elif input_from_user == "1":
        name = input("\nEnter contact name: ")
        phone = input("\nEnter contact phone: ")
        email = input("\nEnter contact email: ")
        
        book.add_contact(name, phone, email)
    
    elif input_from_user == "2":
        name = input("\nEnter contact name: ")
        phone = input("\nEnter contact phone: ")
        email = input("\nEnter contact email: ")
        
        book.updated_contact(name, phone, email)
        
    elif input_from_user == "3":
        print("\n\nList of contacts")
        book.view_contact()
    
    elif input_from_user == "4":
        name = input("\nEnter contact name: ")
        book.delete_contact(name)
        
    else:
        print("Your input is invalid")
        
        
```
